```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open("e_above_hull_info.pkl", "rb") as f:
    eah = pickle.load(f)
```

```python
def get_bg(mpid):
    return eah[mpid]["band_gap"]
```

```python
df = pd.read_pickle('../processing/outputs/post_processed_catlas_data_20220429_vals.pkl')
```

```python
df["band_gap"] = df.bulk_mpid.apply(get_bg)
```

```python
mpids_after = df[(df.e_above_hull <=0.1) & (df["111-like"]) & (df.hit >= 0.1)].bulk_mpid.tolist()
```

```python
mpids_before = df[ (df["111-like"]) & (df.hit >= 0.1)].bulk_mpid.tolist()
```

```python
for mpid in mpids_before:
    if mpid not in mpids_after:
        print(mpid)
```
