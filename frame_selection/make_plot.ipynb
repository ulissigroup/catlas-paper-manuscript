```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_pickle("CO_validation_perframe_gemnet_dT.pkl")
```

```python
absolute_errors = np.array(df.absolute_error.tolist())
```

```python
per_frame_MAE = []
for frame in range(300):
    MAE = sum(absolute_errors[:,frame])/ 81
    per_frame_MAE.append(MAE)
```

```python
print(min(per_frame_MAE), per_frame_MAE.index(min(per_frame_MAE)))
```

```python
plt.plot(range(300), per_frame_MAE)
plt.xlabel("frame number")
plt.ylabel("mean absolute error")
plt.savefig("frame_opt.svg")
plt.savefig("frame_opt.pdf")
```
